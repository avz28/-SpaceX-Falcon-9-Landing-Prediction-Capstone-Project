```python
## Task 1: Mark all launch sites on a map
# Import required libraries
import folium
import pandas as pd
from folium.plugins import MarkerCluster, MousePosition
from folium.features import DivIcon

# Create a DataFrame with launch site names and coordinates
launch_sites = pd.DataFrame({
    'Launch Site': [
        'CCAFS LC-40', 
        'VAFB SLC-4E', 
        'KSC LC-39A', 
        'CCAFS SLC-40'
    ],
    'Lat': [28.562302, 34.632093, 28.573255, 28.561857],
    'Lon': [-80.577356, -120.610829, -80.646895, -80.577366]
})

# Create a folium map centered at the average location
map_center = [launch_sites['Lat'].mean(), launch_sites['Lon'].mean()]
site_map = folium.Map(location=map_center, zoom_start=4)

# Add a MarkerCluster to group the markers
marker_cluster = MarkerCluster().add_to(site_map)

# Add each launch site as a marker to the cluster
for index, row in launch_sites.iterrows():
    folium.Marker(
        location=[row['Lat'], row['Lon']],
        popup=row['Launch Site'],
        icon=folium.Icon(color='blue', icon='rocket', prefix='fa')
    ).add_to(marker_cluster)

# Optional: add mouse position plugin to show lat/lon when you hover
MousePosition().add_to(site_map)

# Display the map (in Jupyter Notebook this will render the map)
site_map
```

```python
# Import required libraries (if not already imported)
import folium
import pandas as pd
from folium.plugins import MarkerCluster, MousePosition
from folium.features import DivIcon

# Launch sites data (reuse your DataFrame if already created)
launch_sites = pd.DataFrame({
    'Launch Site': [
        'CCAFS LC-40', 
        'VAFB SLC-4E', 
        'KSC LC-39A', 
        'CCAFS SLC-40'
    ],
    'Lat': [28.562302, 34.632093, 28.573255, 28.561857],
    'Lon': [-80.577356, -120.610829, -80.646895, -80.577366]
})

# Create the map centered at the average location
map_center = [launch_sites['Lat'].mean(), launch_sites['Lon'].mean()]
site_map = folium.Map(location=map_center, zoom_start=4)

# Add MarkerCluster (optional if you still want clustered markers)
marker_cluster = MarkerCluster().add_to(site_map)

# Add circles and labels for each launch site
for index, row in launch_sites.iterrows():
    coordinate = [row['Lat'], row['Lon']]
    
    # Add a circle with 1000 m radius
    folium.Circle(
        location=coordinate,
        radius=1000,
        color='#000000',
        fill=True,
        fill_opacity=0.4
    ).add_child(folium.Popup(row['Launch Site'])).add_to(site_map)
    
    # Add a text label using DivIcon
    folium.map.Marker(
        coordinate,
        icon=DivIcon(
            icon_size=(100, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12px; color:#d35400;"><b>%s</b></div>' % row['Launch Site'],
        )
    ).add_to(site_map)
    
    # (Optional) also add a regular marker inside marker cluster
    folium.Marker(
        location=coordinate,
        popup=row['Launch Site'],
        icon=folium.Icon(color='blue', icon='rocket', prefix='fa')
    ).add_to(marker_cluster)

# Optional: Add mouse position plugin to display lat/lon when hovering
MousePosition().add_to(site_map)

# Display the map
site_map
```

```python
# Task 2: Mark the success/failed launches for each site on the map
# Import libraries if not yet imported
import folium
import pandas as pd

# Sample data (replace with your actual DataFrame)
spacex_df = pd.DataFrame({
    'Launch Site': ['CCAFS LC-40', 'VAFB SLC-4E', 'KSC LC-39A', 'CCAFS LC-40', 'VAFB SLC-4E'],
    'Lat': [28.562302, 34.632093, 28.573255, 28.562302, 34.632093],
    'Lon': [-80.577356, -120.610829, -80.646895, -80.577356, -120.610829],
    'class': [1, 0, 1, 0, 1]
})

# Create map
site_map = folium.Map(location=[spacex_df['Lat'].mean(), spacex_df['Lon'].mean()], zoom_start=4)

# Add markers
for index, row in spacex_df.iterrows():
    coordinate = [row['Lat'], row['Lon']]
    if row['class'] == 1:
        color = 'green'
        outcome = 'Success'
    else:
        color = 'red'
        outcome = 'Failure'
    
    folium.Marker(
        location=coordinate,
        popup=f"{row['Launch Site']} - {outcome}",
        icon=folium.Icon(color=color, icon='rocket', prefix='fa')
    ).add_to(site_map)

# Display map
site_map
```

```python
from math import radians, cos, sin, sqrt, atan2

def haversine(coord1, coord2):
    # approximate radius of earth in km
    R = 6371.0

    lat1, lon1 = coord1
    lat2, lon2 = coord2

    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)

    a = sin(dlat/2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Example coordinates
launch_site_coord = (28.562302, -80.577356)
coastline_coord = (28.563, -80.567)
city_coord = (28.610, -80.807)
highway_coord = (28.563, -80.570)

# Calculate distances
distance_to_coastline = haversine(launch_site_coord, coastline_coord)
distance_to_city = haversine(launch_site_coord, city_coord)
distance_to_highway = haversine(launch_site_coord, highway_coord)

print(f"Distance to coastline: {distance_to_coastline:.2f} km")
print(f"Distance to city: {distance_to_city:.2f} km")
print(f"Distance to highway: {distance_to_highway:.2f} km")
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
